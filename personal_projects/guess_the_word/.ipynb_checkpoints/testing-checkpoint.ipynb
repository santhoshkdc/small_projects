```python
import re
import random
import os
import pandas as pd
import json
```

```python
%%time
def get_word_list():
    with open('the_time_machine.txt', 'r', encoding='UTF-8') as f:
        f_obj = f.read()
    pattern = re.compile(r"(\b[a-zA-Z]{3,8}\b)")
    matches = re.findall(pattern,f_obj)
    words = list(set(matches))
    with open("word_list.txt", "w+", encoding="UTF-8") as f:
        f.write(str(matches))
    pass
def get_word():
    with open("word_list.json", 'r', encoding='UTF-8') as f:
        words = f.read()
    word = random.choice(words)
    correct_word = word
    print(f"correct word is: {correct_word}")
    split_word = list(word)
    l = len(word)
    idx = list(range(l))
    random.shuffle(idx)
    idx = idx[:(l//2)]
    for i in idx:
        split_word[i] = '_'
    ''.join(split_word)
```
