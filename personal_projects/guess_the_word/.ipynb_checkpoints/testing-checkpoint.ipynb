```python
import random
import requests
import json


class Game_Functions:
    def __init__(self):
        self.length = 0
        self.correct_word = 'XXXX'
        self.dict_ = {}
        self.masked_word = []

    def get_correct_word(self):
        with open("word_list.json", 'r', encoding='UTF-8') as f:
            words = json.load(f)
        correct_word = random.choice(words)
        return correct_word
    
    def clue_definition():
        definition = Game_Functions().dict_[
            'meanings'][0]['definitions'][0]['definition']
        return definition

    def clue_synonyms():
        synonyms = Game_Functions().dict_[
            'meanings'][0]['definitions'][0]['synonyms'][0]
        return synonyms

    def get_response(self,length):
        while True:
            self.correct_word = Game_Functions().get_correct_word()
            if len(self.correct_word) in length:
                pass
            else:
                continue
            print(self.correct_word)
            url = f"https://api.dictionaryapi.dev/api/v2/entries/en/{self.correct_word}"
            response = requests.get(url)
            print('passed')
            try:
                if response.json()[0]['title'] == 'No Definitions Found':
                    continue
                else:
                    break
            except KeyError:
                break
            print('passed')
            try:
                self.dict_ = response.json()[0]
            except KeyError:
                break
            print('passed')
        print(self.correct_word)
    
    def mask_the_word(self):
        word = list(self.correct_word)
        l = len(word)
        idx = list(range(l))
        random.shuffle(idx)
        chunk = l//2
        idx = idx[:chunk]
        for i in idx:
            word[i] = '_'
        self.masked_word = ''.join(word)
        pass
    
    def main_function(self):
        print("It is time to start the game!")
        while True:
            print("Please choose the difficulty level.\n\n*Easy or 1\n*Moderate\
or 2\n*Hard or 3")
            response = input(">> ").lower()
            correct_responses = {
                'e':[3,4,5,6], 'm':[7,8,9,10], 
                'h':[11,12,13,14,15], '1':[3,4,5,6],
                 '2':[7,8,9,10], '3':[11,12,13,14,15]
            }
            if response.startswith(('e', 'm', 'h')):
                self.length = correct_responses[response[0]]
            elif response.isdecimal() and int(response.lower()) in [1,2,3]:
                self.length = correct_responses[response]
            else:
                continue
            break
        length = self.length
        Game_Functions().get_response(length)
        Game_Functions().mask_the_word()
        print("Here is you word to guess:\n>  ",end ='')
        print(self.masked_word)
        response = input('>  ')
        if response == self.correct_word:
            print(f"You guessed it right! Congrats\nThe correct word is '{self.correct_word}'")
        else:
            print(f"Oh, you are wrong! The correct word is '{self.correct_word}'")

g = Game_Functions()
g.main_function()
```
